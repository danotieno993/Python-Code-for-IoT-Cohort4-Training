```python
#Xample 5
'''Create empty list, generate random no. betwn 1 & 30, append the random no. 
generated to the empty list for each iteration of the 20 iterations(len of list)'''
import random
random_list = [] 
for j in range(0, 20): 
    n = random.randint(1, 30)
    random_list.append(n)
print(f'The generated random list is:\n {random_list}')

#iterate over each list element, check whether even or odd and append it to the empty list
evn_list = []
odd_list = []
for elmnt in random_list:
    if elmnt % 2 == 0:
        evn_list.append(elmnt)
    else:
        odd_list.append(elmnt)
print(f'The random list with even elements is:\n {evn_list}')
print(f'The random list with odd elements is:\n {odd_list}')

#swap even list
def swap_even_list(evn_list):
    list_size = len(evn_list)

    #swap 1st & last elements
    first_elmnt = evn_list[0]
    evn_list[0] = evn_list[list_size - 1]
    evn_list[list_size -1] = first_elmnt

    return evn_list

print(f'The swapped random list with even elements is:\n {swap_even_list(evn_list)}')

#swap odd list
def swap_odd_list(odd_list):
    list_size = len(odd_list)

    #swap 1st & last elements
    first_elmnt = odd_list[0]
    odd_list[0] = odd_list[list_size - 1]
    odd_list[list_size - 1] = first_elmnt

    return odd_list

print(f'The swapped random list with odd elements is:\n {swap_odd_list(odd_list)}')
```

```python
# Python program for phone no. missing digits
'''
Write another Python program to find the digits absent in a given mobile number.

'''
def absent_digits(n):
  all_digits = set([0,1,2,3,5,4,6,7,8,9]) #numbers in a mobile phone
  n = set([int(i) for i in n])
  missing_digits = n.symmetric_difference(all_digits)
  missing_digits = sorted(missing_digits)
  return missing_digits
  
phone_num = input("Please enter your phone no.: \n")
while phone_num.isdigit() == False :
    phone_num = input("You Entered Invalid Phone Number...... Please try again \n")
while len(phone_num) < 10:
    phone_num = input("Phone number entered is too short...... Please try again \n")
while len (phone_num) > 10:
    phone_num = input("Phone number entered is too long...... Please try again \n")

#print("The digit(s) absent in your phone number are :\n" + str(absent_digits(phone_num)))
print(f"The digit(s) absent in your phone number are :\n{absent_digits(phone_num)}")
```
